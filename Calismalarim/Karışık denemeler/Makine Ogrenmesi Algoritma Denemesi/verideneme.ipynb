```python
import matplotlib.pyplot as plt
```

```python
def TveCizim(A,B,x:"sınırı belirtin [x,y]"):
    baslangic, bitis = x
    yenix = []
    while baslangic<=bitis:
        yenix.append(baslangic)
        baslangic+=1
    tdegerleri = []
    for i in yenix:
        t = B*i + A
        tdegerleri.append(t)
    plt.plot(yenix,tdegerleri,color='blue')
    plt.show()
```

```python
TveCizim(1,0.5,[-6,6])
```
