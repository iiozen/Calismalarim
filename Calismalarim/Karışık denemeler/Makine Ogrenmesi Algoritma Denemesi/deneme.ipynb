```python
import numpy as np
from sklearn.metrics import r2_score as r2
import pandas as pd
import matplotlib.pyplot as plt
```

```python

veri = pd.read_csv("maaslar_yeni.csv")
veri = veri.drop(['Calisan ID',"unvan"],axis=1)
```

```python
x = veri.iloc[:,:3]
y = veri.iloc[:,3:]
## Aynı işlevler
# y1 = veri[['maas']]
# x1 = veri.drop(['maas'],axis=1,inplace=False)
```

```python
X = x.values
Y = y.values
```

```python
## P Value ile hangi kolonların işe yaradığına bakılacak
## Linear regression dan yardım alınıyor...
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)

plt.scatter(x.index,Y,color='red')
plt.plot(x.index,lin_reg.predict(X),color='blue')
plt.show()

import statsmodels.api as sm

model = sm.OLS(lin_reg.predict(X),X)
model.fit().summary()
```
