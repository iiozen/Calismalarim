```
import time
im=('/content/drive/My Drive/a1.jpeg','/content/drive/My Drive/a2.jpeg','/content/drive/My Drive/a3.jpeg','/content/drive/My Drive/a6.jpeg','/content/drive/My Drive/a5.jpeg')
toplam_nesne=0
toplam_muz=0
toplam_elma=0
toplam_havuc=0
toplam_brokoli=0
z=0

for i in range(5):
  image = read_image_bgr(im[i])
  draw = image.copy()
  draw = cv2.cvtColor(draw, cv2.COLOR_BGR2RGB)
  image = preprocess_image(image)
  image, scale = resize_image(image)
  start = time.time()
  boxes, scores, labels = model.predict_on_batch(np.expand_dims(image, axis=0))
  boxes /= scale
  muz=0
  elma=0
  havuc=0
  brokoli=0
  nesne=0 
  for box, score, label in zip(boxes[0], scores[0], labels[0]):
    if score < 0.5:
      break
    color = label_color(label)
    
    b = box.astype(int)
    draw_box(draw, b, color=color)
    
    caption = "{} {:.3f}".format(labels_to_names[label], score)
    draw_caption(draw, b, caption)
    if label==50:
      brokoli=brokoli+1
      toplam_brokoli=toplam_brokoli+1
      toplam_nesne=toplam_nesne+1
      nesne=nesne+1
    if label==51:
      nesne=nesne+1
      toplam_nesne=toplam_nesne+1
      toplam_havuc=toplam_havuc+1
      havuc=havuc+1
    if label==47:
      nesne=nesne+1
      toplam_nesne=toplam_nesne+1
      toplam_elma=toplam_elma+1
      elma=elma+1
    if label==46:
      nesne=nesne+1
      toplam_nesne=toplam_nesne+1
      toplam_muz=toplam_muz+1
      muz=muz+1
  plt.xticks([])
  plt.yticks([])
  plt.imshow(draw)
  plt.xlabel('Nesne=%i, Muz=%i, Elma=%i, Havu=%i, Brokoli=%i.'%(nesne,muz,elma,havuc,brokoli),fontsize=15,color='cyan')
  plt.show()
  z=z+1
q='\n\n\nBtn Fotoraflardaki Toplam Nesne=%i, Toplam Muz=%i, Toplam Elma=%i, Toplam Havu=%i, Toplam Brokoli=%i.'%(toplam_nesne,toplam_muz,toplam_elma,toplam_havuc,toplam_brokoli)
print(q)

```
