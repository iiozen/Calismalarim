```
from keras.datasets import fashion_mnist
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix as cm
import numpy as np
```

```
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```
nsamples, nx, ny = x_train.shape
x_train1 = x_train.reshape((nsamples,nx*ny))
d_tree1 = DecisionTreeClassifier(max_depth = 50, random_state=42)
d_tree2 = DecisionTreeClassifier(max_depth = 30, random_state=42)
d_tree3 = DecisionTreeClassifier(max_depth = 70, random_state=42)
d_tree4 = DecisionTreeClassifier(criterion='entropy',max_depth = 50, random_state=42)
elli=d_tree1.fit(x_train1, y_train)
otuz=d_tree2.fit(x_train1, y_train)
yetmis=d_tree3.fit(x_train1, y_train)
ellientr=d_tree4.fit(x_train1, y_train)
```

```
from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix as cm
from sklearn import datasets, neighbors, linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline 
import seaborn as sns
nsamples, nx, ny = x_test.shape
x_test1 = x_test.reshape((nsamples,nx*ny))
predictions1 = d_tree1.predict(x_test1)
predictions2 = d_tree2.predict(x_test1)
predictions3 = d_tree3.predict(x_test1)
predictions4 = d_tree4.predict(x_test1)
score1 = round(accuracy_score(y_test, predictions1), 3)
score2 = round(accuracy_score(y_test, predictions2), 3)
score3 = round(accuracy_score(y_test, predictions3), 3)
score4 = round(accuracy_score(y_test, predictions4), 3)
cm1 = cm(y_test, predictions1)
cm2 = cm(y_test, predictions2)
cm3 = cm(y_test, predictions3)
cm4 = cm(y_test, predictions4)
sns.heatmap(cm1,annot=True,fmt=".0f")
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Accuracy Score: {0}'.format(score1), size = 15)
plt.show()
sns.heatmap(cm2, annot=True, fmt=".0f")
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Accuracy Score: {0}'.format(score2), size = 15)
plt.show()
sns.heatmap(cm3, annot=True, fmt=".0f")
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Accuracy Score: {0}'.format(score3), size = 15)
plt.show()
sns.heatmap(cm4, annot=True, fmt=".0f")
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Accuracy Score: {0}'.format(score4), size = 15)
plt.show()
```

```
basarili_sonuc=[score1],[score2],[score3],[score4]
basarilicol=np.argmax(np.max(basarili_sonuc, axis=1))
sonuclar=['Derinlik Bilgisi 50 olan eğitim','Derinlik Bilgisi 30 olan eğitim','Derinlik Bilgisi 70 olan eğitim','Derinlik Bilgisi 50 ve Criterion=Entropy olan eğitim']
print('En başarılı sonuç veren eğitim: ',sonuclar[basarilicol])
```
