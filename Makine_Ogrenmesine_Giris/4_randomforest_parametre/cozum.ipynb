```
from sklearn import tree
import keras
from keras.datasets import fashion_mnist
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score,precision_score,f1_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import classification_report
```

```
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```
NuOfImages_x_train= x_train.shape[0]
NuOfImages_x_test= x_test.shape[0]
WidthOfImages= x_train.shape[1]
x_train=np.reshape(x_train,(NuOfImages_x_train,WidthOfImages*WidthOfImages))
x_test=np.reshape(x_test,(NuOfImages_x_test,WidthOfImages*WidthOfImages))
```

```
clf = RandomForestClassifier(max_depth=3,n_estimators=10,random_state=0)
clf.fit(x_train, y_train)
```

```
predictions_train=clf.predict(x_train)
predictions_test=clf.predict(x_test)
(acc_train),(acc_test)=round((accuracy_score(y_train,predictions_train)),3),round((accuracy_score(y_test,predictions_test)),3)
(recall_train),(recall_test)=round((recall_score(y_train,predictions_train,average='macro')),3),round((recall_score(y_test,predictions_test,average='macro')),3)
(precision_train),(precision_test)=round((precision_score(y_train,predictions_train,average='macro')),3),round((precision_score(y_test,predictions_test,average='macro')),3)
(f1_train),(f1_test)=round((f1_score(y_train,predictions_train,average='macro')),3),round((f1_score(y_test,predictions_test,average='macro')),3)
print('Eğitim veri setinde accuracy=',acc_train,'recall=',recall_train,'precision=',precision_train,'f1=',f1_train,'\n\n'
,'Test veri setinde accuracy=',acc_test,'recall=',recall_test,'precision=',precision_test,'f1=',f1_test)
```

```
# ÖZNİTELİĞİN SAYISAL OLARAK SIRASINDAKİ ÖNEM DEĞERİNE GÖRE OLUŞTURULMUŞ GRAFİK
importance = clf.feature_importances_
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```

```
from sklearn.model_selection import GridSearchCV
parameters={'n_estimators':[10,50],'min_samples_leaf':[2,3],'max_depth':[3,10]}
scores=['precision','recall','f1']
for score in scores:
  print("# Tuning hyper-parameters for %s"%score)
  print()
  klf=GridSearchCV(clf,parameters,scoring='%s_macro'%score)
  klf.fit(x_train,y_train)
  print("Best parameters set found on development set:")
  print()
  print(klf.best_params_)
  print()
  print("Grid Scores on development set:")
  print()
  means = klf.cv_results_['mean_test_score']
  stds = klf.cv_results_['std_test_score']
  for mean, std, params in zip(means, stds, klf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))
  print()

  print("Detailed classification report:")
  print()
  print("The model is trained on the full development set.")
  print("The scores are computed on the full evaluation set.")
  print()
  y_true, y_pred = y_test, klf.predict(x_test)
  print(classification_report(y_true, y_pred))
  print()


```
